```python
import networkx as nx
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Hei']
```

```python
edges = [
    ("华为", "手机"), 
    ("苹果", "手机"), 
    ("小米", "手机"), 
    ("华为", "高端市场"), 
    ("苹果", "高端市场"), 
    ("小米", "低端市场")
]
```

```python
G = nx.Graph()
G.add_edges_from(edges)
plt.figure(figsize=(8,6))
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=2000, node_color="#3b82f6")
nx.draw_networkx_edges(G, pos, edge_color="#475569")
nx.draw_networkx_labels(G, pos, font_color="white", font_size=10)
plt.show()
```

```python
preds = nx.jaccard_coefficient(G)
```

```python
companies = ["华为", "苹果", "小米"]
```

```python
for n1, n2, p in list(preds):
    if n1 in companies and n2 in companies:
        print("{0}与{1}的相似度为：{2:.2f}".format(n1, n2, p))
```
